```python
from flask_ngrok import run_with_ngrok
from flask import Flask, request, render_template
import pickle
import numpy as np
```

```python
#running the flask app
app = Flask(__name__)

#start ngrok when the app is run
run_with_ngrok(app)

#load model
model = pickle.load(open('model_cem.pkl', 'rb'))

@app.route('/')
def home():
  return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
  sid = request.form.get("sid").upper()
  sname = request.form.get("sname").upper()
  branch = request.form.get("branch").upper()
  degree = request.form.get("degree").upper()
  
  features = [x for x in request.form.values()]
  final_features = (np.array(features[4:], dtype=float)).reshape(1, -1)
       
  prediction = model.predict(final_features)
  
  if prediction[0] == 0:
    placement_status = "PLACED"
  else:
    placement_status = "NOT PLACED"

  output = (sid, sname, branch, degree, placement_status)
  
  return render_template('result.html', 
                         sid='{}'.format(output[0]), 
                         sname='{}'.format(output[1]), 
                         branch='{}'.format(output[2]), 
                         degree='{}'.format(output[3]), 
                         placement_status='{}'.format(output[4]))
  
  #return "{}".format(tuple(final_features))

app.run()
```
